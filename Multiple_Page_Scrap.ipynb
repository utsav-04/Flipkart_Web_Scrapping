```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
httpObject=urlopen("https://www.flipkart.com/q/best-laptops-under-rs-50000")
webdata=httpObject.read()
print(webdata)
```

```python
soupdata=soup(webdata)
```

```python
page_link=soupdata.findAll('a',{'class':'ge-49M'})
domain='https://www.flipkart.com/q/best-laptops-under-rs-50000?page='

for i in range(2,42):
  link=domain+str(i)
  page_data=urlopen(link)
  webdata1=page_data.read()
  webdata +=webdata1
```

```python
soupdata=soup(webdata,'html.parser')
```

```python
containers=soupdata.findAll('div',{'class':'_2kHMtA'})
print(len(containers))
```

```python
for container in containers:
  product=container.findAll('div',{'class':'_4rR01T'})
  print(product[0].text)
```

```python
for container in containers:

  product=container.find('div',{'class':'_4rR01T'})
  ProductName=product.text.split('-')[0].strip()
  print(ProductName)
```

```python
for container in containers:
  info=container.findAll('li',{'class':'rgWa7D'})
  Processor=info[0].text
  Ram=info[1].text
  Storage=info[3].text
  print(Processor,Ram,Storage)
```

```python
for container in containers:

  Rating=container.find('span',{'class':'_2_R_DZ'})
  try:
    ratRev=re.findall('\d+,?\d*',Rating.text)
    Ratings=ratRev[0]
    Reviews=ratRev[1]
  except:
    Ratings=0
    Reviews=0
  print(Ratings, Reviews)
```

```python
for container in containers:
  currentprice=container.find('div',{'class':'_30jeq3 _1_WHN1'})
  try:
    CurrentPrice=currentprice.text
  except:
    CurrentPrice=0
  print(CurrentPrice)
```

```python
for container in containers:
  mrp = container.find('div',{'class':'_3I9_wc _27UcVY'})
  try:
    MRP=mrp.text
  except:
    MRP=0
  print(MRP)
```

```python
for container in containers:
  star=container.find('div',{'class':'_3LWZlK'})
  try:
    Stars=star.text
  except:
    Stars=0
  print(Stars)
```

```python
f=open('Flipkart_scrap.csv','wb')
f.write('ProductName,Stars,Ratings,Reviews,MRP,CurrentPrice,Processor,Ram,Storage,ImageUrl\n'.encode())
import re
for container in containers:
  product=container.find('div',{'class':'_4rR01T'})
  ProductName=product.text.split('-')[0].strip()

  star=container.find('div',{'class':'_3LWZlK'})
  try:
    Stars=star.text
  except:
    Stars=0

  Rating=container.find('span',{'class':'_2_R_DZ'})
  try:
    ratRev=re.findall('\d+,?\d*',Rating.text)
    Ratings=ratRev[0].replace(',','')
    Reviews=ratRev[1].replace(',','')
  except:
    Ratings=0
    Reviews=0

  currentprice=container.find('div',{'class':'_30jeq3 _1_WHN1'})
  try:
    CurrentPrice=currentprice.text.replace(',','').replace('₹','')
  except:
    CurrentPrice=0

  mrp = container.find('div',{'class':'_3I9_wc _27UcVY'})
  try:
    MRP=mrp.text.replace(',','').replace('₹','')
  except:
    MRP=0

  info=container.findAll('li',{'class':'rgWa7D'})
  Processor=info[0].text
  Ram=info[1].text
  Storage=info[3].text

  Image=container.img
  ImageUrl=Image.get('src')
  print(ProductName,Stars,Ratings,Reviews,MRP,CurrentPrice,Processor,Ram,Storage,ImageUrl)
  f.write(f"{ProductName},{Stars},{Ratings},{Reviews},{MRP},{CurrentPrice},{Processor},{Ram},{Storage},{ImageUrl}\n".encode())
  print('\n')
f.close()
```
